```python
string =input("enter sentence : ")

words=string.split()
unique_words=[]

for i in words:
  if i not in unique_words:
    unique_words.append(i)

print(f"\n Unique words:{unique_words}\n")
print(f"Unique words count:{len(unique_words)}")
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

max_sal=0
name=""
for i in employees:
  if employees[i]>max_sal:
    max_sal=employees[i]
    name=i

print(f"Highest salary: {name} - {max_sal}")
```

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")

total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
text = "pythopn"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```
