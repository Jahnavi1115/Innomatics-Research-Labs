```python
'''
Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ‚â• ‚Çπ50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry


'''

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ‚Çπ"))

            if amount < 50:
                print("‚ùå Recharge amount must be at least ‚Çπ50. Try again.\n")
                continue

            if amount not in valid_plans:
                print("‚ùå Invalid plan selected.")
                print("Available plans: 199, 299, 399, 599\n")
                continue

            print("‚úÖ Recharge successful!")
            break

        except ValueError:
            print("‚ùå Please enter a valid numeric amount.\n")

validate_recharge()
```

```python
'''
Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary


'''

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert ‚ö†Ô∏è (Stock: {stock})")
        else:
            print(f"{product}: Stock OK ‚úÖ (Stock: {stock})")


inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 12
}

check_inventory(inventory)
```

```python
'''
 Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ‚â• 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function
'''

def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average:.2f}")

    if average >= 50:
        print("Result: Pass ‚úÖ")
        return "Pass"
    else:
        print("Result: Fail ‚ùå")
        return "Fail"


student_marks = [60, 45, 70, 55]

process_result(student_marks)
```

```python

'''
Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ‚Çπ50


‚Çπ12 per km


Peak hour ‚Üí 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry
'''

def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    total_fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        total_fare += total_fare * 0.25

    return total_fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(distance, peak)

        print(f"Total Fare: ‚Çπ{fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator üöñ")
            break

    except ValueError:
        print("Please enter valid numeric distance.\n")
```

```python
'''
Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ‚â• 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement
'''

def check_attendance(attendance):
    present_days = 0
    total_days = len(attendance)

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_record = ["P", "P", "A", "P", "P", "A", "P"]

result = check_attendance(attendance_record)
print("Status:", result)
```

```python
'''
Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function
'''

def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password ‚ùå (Minimum length is 8)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password ‚úÖ"
    else:
        return "Weak Password ‚ùå (Must include digit and @/#/$)"


user_password = input("Enter password: ")
result = check_password(user_password)
print(result)
```
